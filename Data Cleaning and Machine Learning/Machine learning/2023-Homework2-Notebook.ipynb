# Section 0 - Reclean Data

In this section we begin from sratch using an unclean data set.  We will take similar steps as in homework 1, meaning for the mostpart our cleaned data will look the same as in homework 1.  However there will be some key differences.  Under further examination the choice has been made to keep some features and change the data type for other features.  The difference is highlighted below:

1. Update case month to unix timestamp.  In homework 1 we considered this catagorical data but we swap to a continuous unix timestamp. (could use datetime but timestamp is chosen for manipulation purposes)

2. In homework 1 the choice was made to drop ICU and Underlying_conditions column.  We choose to keep them in this homework as we are now making the assumption that all ICU cases and Underlying conditions would be well recorded.  Such that if the data is missing, we count it as 'no' case.

3. Choose to drop case positive specimen interval and case onset interval.  These data columns were highlighted in homework 1 as problematic.  Categortical data is massively skewed towards the bucket value 0 and outliers are physically impossible implying the data column contains bad data.  The ammount of processing necessary to extract correlated information from this column is too great.  The data processing biases the data column towards the expectation of the data scientist and therefore the columns are unreliable.

4. ICU_yn and Hosp_yn 'Missing' value is replaced with 'No'

**Justification for dropping 'case_positive_specimen_interval' and 'case_onset_interval':**  The difficulty with extracting data from these columns is that the data scientist has a preconceived expectation on what to expect from these columns.  The application of data mappings can be good in some cases to extract more easily understandable metrics from data.  However in this case the procedures which have to be applied in order to extract correlated data fundamentally change the underlying distributions.  This implies that the procedures chosen to map the data in homework 1 were actually biased in their selection of data points which fit the narrative of the data scientist.  Therefore in this homework we make the choice to drop these columns.

**Justification for ICU_yn and Hosp_yn columns:**  We define these two columns as healthservice columns.  Healthservice columns represent a column which has high likely hood of being filled in if the patient has been admitted to some health service provider such as a hospital.  The expectation is that these columns would be filled in the case that a patient was admitted to these services.  Therefore the fact that the column entries are missing implies the patient was not admitted to these services.

```python
#Import the required packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load and display size of data

df = pd.read_csv('data/covid19-cdc-15463152.csv', keep_default_na=True, delimiter=',', skipinitialspace=True)
print(f"The data has {df.shape[1]} Columns and {df.shape[0]} Rows")
```

```python
# Change state and county fips code to categorical data type

df["state_fips_code"] = df["state_fips_code"].dropna().astype('Int64')
df["state_fips_code"] = df["state_fips_code"].astype('category')
df["county_fips_code"] = df["county_fips_code"].dropna().astype('Int64')
df["county_fips_code"] = df["county_fips_code"].astype('category')
```

```python
# Convert casemonth to Unix timestamp

df['case_month'] = pd.to_datetime(df['case_month'])
df['case_month'] = df['case_month'].astype('int64') // 10**9
```

```python
# Drop columns

df = df.drop(['process', 'exposure_yn'], axis=1)
df = df.drop(['case_positive_specimen_interval', 'case_onset_interval'], axis=1)
```

```python
# Change object columns to category type

object_columns = df.select_dtypes(['object']).columns
for column in object_columns:
    df[column] = df[column].astype('category')
```

```python
# Remove data types

df['symptom_status'] = df['symptom_status'].replace(['Unknown', 'Missing'], 'Missing')
```

```python
# Ensure 'Missing' is a category and replace none cats with 'Missing'

category_columns = df.select_dtypes(['category']).columns
for column in category_columns:
    if 'Missing' not in df[column].cat.categories:
        df[column] = df[column].cat.add_categories(['Missing'])
    df[column] = df[column].replace('Unknown', 'Missing')
        
df[category_columns] = df[category_columns].fillna('Missing')
```

```python
# Replace Missing with no for two healthservice columns

df['hosp_yn'] = df['hosp_yn'].replace('Missing','No')
df['icu_yn'] = df['icu_yn'].replace('Missing','No')
```

```python
# Data snapshot


df_null = df[category_columns].isnull().sum()
df_miss_unk = df[category_columns].isin(["Missing","Unknown"]).sum()
category_columns_perc_missing  = 100*(df_null+df_miss_unk)/df.shape[0]
df_table_categ = df[category_columns].describe().T
df_category_perc_missing = pd.DataFrame(category_columns_perc_missing, columns=['%missing'])
category_columns_card = df[category_columns].nunique()
df_category_card = pd.DataFrame(category_columns_card, columns=['card'])
df_category_columns_data_quality_report_table = pd.concat([df_table_categ, df_category_perc_missing, df_category_card], axis=1)

df_category_columns_data_quality_report_table
```

# Section 2 - Predictive Modeling: Linear Regression


## 2.1 - Training the Linear Regression Model


Here we train the linear regression model using the sklearn module.  We select features: **'case_month', 'icu_yn', 'hosp_yn', 'underlying_conditions_yn', 'age_group', and 'sex'** to train our model.  For our categorical features we use pandas.get_dummies method to substitute dummy variables for our categories.  We then append our continuous month feature.  We then train our model.  In our target feature we represent death_yn: yes as 1 and death_yn: no as 0.

```python
#Import the required packages

from patsy import dmatrices
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_score, KFold, cross_val_predict
```

```python
# Train the Linear regression using dummy variables for categorical features

features = ['case_month', 'icu_yn', 'hosp_yn', 'underlying_conditions_yn', 'age_group', 'sex']
target = 'death_yn'
continuous_data = train_df['case_month']
X_categorical = pd.get_dummies(train_df[features].drop('case_month', axis=1), drop_first=True)
X = pd.concat([continuous_data, X_categorical], axis=1)
y = train_df[target].apply(lambda x: 1 if x == 'Yes' else 0)

linreg = LinearRegression().fit(X, y)
```

1. underlying_conditions_yn (Yes = -0.1239): Interestingly, having underlying conditions also slightly decreases the odds of the outcome, which might require further exploration to understand if certain conditions are being effectively managed or are less impactful.

1. 65+ years (3.2859): The oldest age group has the highest positive coefficient, which starkly highlights the increased risk associated with this age demographic.
