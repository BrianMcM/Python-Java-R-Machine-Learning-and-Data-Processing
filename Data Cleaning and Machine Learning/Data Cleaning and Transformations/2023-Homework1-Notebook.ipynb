### Ratio of deaths depending on age

Below is a split bar plot of how the change of death is related to age of the testee.  It is quite clear to see that as age increases the liklihood of death does also.

```python

# Define the order of categories for the x-axis
DeathLabels = ['No', 'Yes']

# Group by 'age_group'
grouped = df_clean.groupby('age_group')

# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(8, 8))
fig.tight_layout()

plt.subplots_adjust(hspace=0.4, wspace=0.4)
# Iterate over groups and plot
i = 0
j = 0
for name, group in grouped:
    # Calculate value counts for 'death_yn' column and ensure the order of categories
    counts = df_clean[df_clean['age_group'] == name]['death_yn'].value_counts(dropna=False).reindex(DeathLabels)
    # Plot bar chart for each age group
    counts.plot(kind='bar', title=name, ax=axes[j, i])
    axes[j, i].set_xlabel('Death')
    axes[j, i].set_ylabel('Count')
    i += 1
    if i == 2:
        j += 1
        i = 0

plt.show()
```

We will use state population and month of the year to create a feature which tries to capture some of the impact on healthcare system.
```python
df_clean['case_year'] = pd.to_datetime(df_clean['case_month']).dt.year
sp.dtypes
sp['State Fips Code'] = sp['State Fips Code'].astype('int32')
```

```python
#function to add population data of states to rows

def get_population(df, year, fips_code):
    try:
        # Filter the DataFrame based on the year and FIPS code
        population = df.loc[df['State Fips Code'] == fips_code, str(year)].iloc[0]
        return population
    except IndexError:
        return np.nan



# Apply the get_population function to create the 'state_pop' column
df_clean['state_pop'] = df_clean.apply(lambda row: get_population(sp, row['case_year'], row['state_fips_code']), axis=1)
```
